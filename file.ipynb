```python
data = xr.open_dataset('./data/plevel_data.nc')
surface_data = xr.open_dataset('./data/surface_data.nc')
#print(data)
divergence = data['d'].sel(pressure_level=900, valid_time='2018-09-12T00:00:00', method='nearest')
print(divergence.values)

print("--------------------------------")

divergence2 = data['d'].sel(pressure_level=900, valid_time='2018-09-12T06:00:00', method='nearest')
print(divergence2.values)

print("--------------------------------")

divergence3 = data['d'].sel(pressure_level=900, valid_time='2018-09-12T12:00:00', method='nearest')
print(divergence3.values)

print("--------------------------------")

divergence4 = data['d'].sel(pressure_level=900, valid_time='2018-09-12T18:00:00', method='nearest')
print(divergence4.values)
```

```python
"""
sst = sea surface temperature
u10, v10 = 10m surface winds
d = divergence
pv = potential vorticity
vo = relative vorticity
w = vertical velocity
u, v = wind to compute shear
r = relative humidity
"""
full_data = xr.merge([data, surface_data])
print(full_data)

print(full_data.variables)
```

```python
"""
Calculating verticle wind shear 200-700 hPa
"""
u_upper = full_data['u'].sel(pressure_level=200)
u_lower = full_data['u'].sel(pressure_level=700, method='nearest')
v_upper = full_data['v'].sel(pressure_level=200)
v_lower = full_data['v'].sel(pressure_level=700, method='nearest')

verticle_shear = np.sqrt((u_upper - u_lower)**2 + (v_upper - v_lower)**2)
print(verticle_shear)
```
